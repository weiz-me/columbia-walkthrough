```python
#Establish path
import os
import datetime
path=os.path.abspath('')
#print(path)

#folder=input("Folder name:\n")
pathfile=path+"\\images"
os.chdir(pathfile)
```

```python
#rename
count = 1
date = "03162023"

for filename in os.listdir():
        t = os.path.getmtime(filename)
        v = datetime.datetime.fromtimestamp(t)
        x = v.strftime('%Y%m%d-%H%M%S')
        os.rename(filename, x +".jpg")

os.chdir(pathfile)
for file in os.listdir():
    current_file_name = file
    new_file_name = f"{date}-panorama{count}.jpg"
    count +=1

    # Rename the file
    os.rename(current_file_name, new_file_name)
```

```python
#Establish path
import os
path=os.path.abspath('')
#print(path)
date = "03162023"

#folder=input("Folder name:\n")
os.chdir(pathfile)

#print script part 1

script1='const viewer = new PANOLENS.Viewer( );'
urls=[]
for file in os.listdir():

    pano=file[file.find("panorama")-len(file):-4]
    script1=script1+f'\nconst {pano} = new PANOLENS.ImagePanorama( "images/{file}" );'
    script1=script1+f'\nviewer.add( {pano});'
    urls.append(f"../images/{file}")
    
print(script1)
#print(urls)



#print script part 2 pano using excel
os.chdir(path)
import openpyxl
from pathlib import Path

wb_obj = openpyxl.load_workbook("../pano.xlsx") 

# Read the active sheet:
sheet = wb_obj.active
row=sheet.max_row
script2=""
for num in range(4,row+1):
    script2=script2+"\n"+sheet[f"A{num}"].value

for num in range(4,row+1):
    script2=script2+"\n"+sheet[f"B{num}"].value

print("\n\n\n"+script2+"\n\n\n")






#print script part 3 GPS


    
#compile for loop
import PIL.Image
import PIL.ExifTags
import os 
#print(pathimg)
#print(urls)

lat=["no p0"]
lon=["no p0"]

for url in urls:
    img = PIL.Image.open(url)
    exif_data = img._getexif()

    exif = {
        PIL.ExifTags.TAGS[k]: v
        for k, v in img._getexif().items()
        if k in PIL.ExifTags.TAGS
    }
    #print(exif)
    
    if exif["GPSInfo"][1] =='N':
        latdata = str(exif["GPSInfo"][2][0])
    if exif["GPSInfo"][1] =='S':
        latdata = "-"+str(exif["GPSInfo"][2][0])
    lat.append(latdata)
    
    if exif["GPSInfo"][3] =='E':
        londata = str(exif["GPSInfo"][4][0])
    if exif["GPSInfo"][3] =='W':
        londata = "-"+str(exif["GPSInfo"][4][0])
    lon.append(londata)
    #print(url)
    img.close()
    #print()
    
#HMC
counting=1
for url in urls:
    num=url[url.find("panorama")-len(url)+len("panorama"):-4]
    num = int(num)
    print(f"var marker{num} = L.marker([{lat[counting]},{lon[counting]}]).on('click', onClick{num}).addTo(map);\nfunction onClick{num}(e) "+"{"+'\ndocument.getElementById("map").style="width: 27%; height: 30%; position: absolute;  bottom: 1%; left: 1%; max-width: 350px; z-index: 1001" ;\nmap.setView(['+f"{lat[counting]},{lon[counting]}], 19);\nsetTimeout(function()"+"{"+'map.invalidateSize()}, 600);\nviewer.setPanorama( panorama'+str(num)+");}")
    counting = counting +1
```
